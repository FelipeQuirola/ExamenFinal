```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
from scipy.optimize import newton
```

```python
def P(x):
    return x**5 - 6*x**4 + 2*x**3 + 20*x**2 - 27*x + 10
```

```python
x = np.linspace(-5, 7, 400)
y = P(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r'$P(x) = x^5 - 6x^4 + 2x^3 + 20x^2 - 27x + 10$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title("Gr√°fico del polinomio $P(x)$")
plt.xlabel("x")
plt.ylabel("P(x)")
plt.ylim(-250, 250)
plt.grid(True)
plt.legend()
plt.show()
```

```python
def P_prime(x):
    return 5*x**4 - 24*x**3 + 6*x**2 + 40*x - 27

initial_guesses = [-10, -5, 0, 2, 5, 10]
roots = []

for guess in initial_guesses:
    root = newton(P, x0=guess, fprime=P_prime)
    roots.append(root)

unique_roots = np.unique(np.round(roots, decimals=8))

print(unique_roots)
```
